```python
import matplotlib.pyplot as plt

# Calculate missing value percentages
missing_percent = (df.isnull().sum() / len(df) * 100)
missing_percent =missing_percent [missing_percent >2]
# Pie chart with percentages
plt.figure(figsize=(8, 8))
plt.pie(
    missing_percent,
    labels=missing_percent.index,
    autopct='%1.1f%%',  # shows percentage on the slices
    startangle=140,
    colors=plt.cm.Paired.colors
)
plt.title('Percentage of Missing Values by Column', fontsize=14)
plt.axis('equal')  # Equal aspect ratio ensures pie is circular
plt.show()

```

```python
numerical_cols = ["Population(2022)", "Area", "% of World", "Density(km2)"]

for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())
```
